## First whyyoulyingchallenge

```python
tag_name_two='whyyoulyingchallenge'
```

```python
from tiktokapipy.async_api import AsyncTikTokAPI

async def get_challenge_videos_scroll_down_tag_name_two():
    # scroll down for 2.5 seconds when making requests
    async with AsyncTikTokAPI(scroll_down_time=3) as api:
        #initialise keys of dictionary
        dict_titles=['video_author',
                 'video_challenges',
                 'video_comments',
                 'time_of_creation',
                 'video_creator',
                 'caption',
                 'video_image_post',
                 'video_music',
                 'video_music_title',
                 'num_comments',
                 'num_likes',
                 'num_views',
                 'video_music_title',
                 'num_shares',
                 'video_tags',
                 'video_data',
                 'video_id']

        #initialise empty list which will contain all new video entries
        global all_video_entries
        all_video_entries=[]
        challenge = await api.challenge(tag_name_two)
        
 
        async for video in challenge.videos:
            
            global whyyoulyingchallenge_list
            whyyoulyingchallenge_list=[]
            
            video_author=video.author
            video_challenges=video.challenges
            video_comments=video.comments
            time_of_creation = video.create_time
            video_creator=video.creator
            caption = video.desc
            video_image_post=video.image_post
            video_music=video.music
            video_music_title=video.music.title
            num_comments = video.stats.comment_count
            num_likes = video.stats.digg_count
            num_views = video.stats.play_count
            num_shares = video.stats.share_count
            video_tags=video.tags
            video_data=video.video
            video_id=video.id            

            
            
            whyyoulyingchallenge_list.append(video_author)
            whyyoulyingchallenge_list.append(video_challenges)
            whyyoulyingchallenge_list.append(video_comments)
            whyyoulyingchallenge_list.append(time_of_creation)
            whyyoulyingchallenge_list.append(video_creator)
            whyyoulyingchallenge_list.append(caption)
            whyyoulyingchallenge_list.append(video_image_post)
            whyyoulyingchallenge_list.append(video_music)
            whyyoulyingchallenge_list.append(video_music_title)
            whyyoulyingchallenge_list.append(num_comments)
            whyyoulyingchallenge_list.append(num_likes)
            whyyoulyingchallenge_list.append(num_views)
            whyyoulyingchallenge_list.append(num_shares)
            whyyoulyingchallenge_list.append(video_tags)
            whyyoulyingchallenge_list.append(video_data)
            whyyoulyingchallenge_list.append(video_id)

#             whyyoulyingchallenge_list.append(video_link)


            
            list_of_one_video_detail = dict(zip(dict_titles,whyyoulyingchallenge_list))

            all_video_entries.append(list_of_one_video_detail)
        return all_video_entries

    # instead of list grab detail put in dictionary then put dictionary in list
```

```python
tag_name='whyareyoulyingchallenge'
```

```python
from tiktokapipy.async_api import AsyncTikTokAPI

async def get_challenge_videos_scroll_down():
    # scroll down for 2.5 seconds when making requests
    async with AsyncTikTokAPI(scroll_down_time=3) as api:
        #initialise keys of dictionary
        dict_titles=['video_author',
                 'video_challenges',
                 'video_comments',
                 'time_of_creation',
                 'video_creator',
                 'caption',
                 'video_image_post',
                 'video_music',
                 'video_music_title',
                 'num_comments',
                 'num_likes',
                 'num_views',
                 'video_music_title',
                 'num_shares',
                 'video_tags',
                 'video_data']

        #initialise empty list which will contain all new video entries
        global all_video_entries_whyare
        all_video_entries_whyare=[]
        challenge = await api.challenge(tag_name)
        

        
        async for video in challenge.videos:
            
            global whyareyoulyingchallenge_list
            whyareyoulyingchallenge_list=[]
        
            video_author=video.author
            video_challenges=video.challenges
            video_comments=video.comments
            time_of_creation = video.create_time
            video_creator=video.creator
            caption = video.desc
            video_image_post=video.image_post
            video_music=video.music
            video_music_title=video.music.title
            num_comments = video.stats.comment_count
            num_likes = video.stats.digg_count
            num_views = video.stats.play_count
            num_shares = video.stats.share_count
            video_tags=video.tags
            video_data=video.video
            video_link=video.video_link

            
            
            whyyoulyingchallenge_list.append(video_author)
            whyyoulyingchallenge_list.append(video_challenges)
            whyyoulyingchallenge_list.append(video_comments)
            whyyoulyingchallenge_list.append(time_of_creation)
            whyyoulyingchallenge_list.append(video_creator)
            whyyoulyingchallenge_list.append(caption)
            whyyoulyingchallenge_list.append(video_image_post)
            whyyoulyingchallenge_list.append(video_music)
            whyyoulyingchallenge_list.append(video_music_title)
            whyyoulyingchallenge_list.append(num_comments)
            whyyoulyingchallenge_list.append(num_likes)
            whyyoulyingchallenge_list.append(num_views)
            whyyoulyingchallenge_list.append(num_shares)
            whyyoulyingchallenge_list.append(video_tags)
            whyyoulyingchallenge_list.append(video_data)
            whyyoulyingchallenge_list.append(video_link)


            
            list_of_one_video_detail_whyare = dict(zip(dict_titles,whyareyoulyingchallenge_list))

            all_video_entries_whyare.append(list_of_one_video_detail_whyare)
        return all_video_entries_whyare

    # instead of list grab detail put in dictionary then put dictionary in list
```
